```python
# Importing the Libraries
import pandas as pd
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
import time
import string
from datetime import date
from datetime import datetime
start_page=0
end_page=34
# Giving the name of the website which we have to scrap
def data_scientist(url):   # Scraping data analyst profile for delhi location
    #Loading Driver
    driver=webdriver.Chrome(r"C:\Users\dell\Downloads\ML.data\chromedriver_win32\chromedriver.exe")
    # Getting the URL
    driver.get(url)
    # we have to scrap the below attached details
    Job_Name=[]
    Job_Location=[]
    Company_Name=[]
    Salary=[]
    Job_Desc=[]
    Post_Days=[]
    crawl_timestamp=[]
    crawl_date=[]
    for page in range(start_page,end_page+1):
        name=driver.find_elements_by_xpath("//h2[@class='title']")
        for n in name:
            n_name=n.text
            Job_Name.append(n_name)
            JN=pd.DataFrame(Job_Name)
            JN=JN.rename({0: 'Job Name'},axis=1)
        location=driver.find_elements_by_xpath("//span[@class='location accessible-contrast-color-location']")
        for ln in location:
            ln_name=ln.text
            Job_Location.append(ln_name)
            LN=pd.DataFrame(Job_Location)
            LN=LN.rename({0: 'Job Location'},axis=1)
        company=driver.find_elements_by_xpath("//div[@class='sjcl']/div[1]")
        for c in company:
            c_name=c.text
            Company_Name.append(c_name)
            CN=pd.DataFrame(Company_Name)
            CN=CN.rename({0: 'Company Name'},axis=1)
        salary=driver.find_elements_by_xpath("//span[@class='salary no-wrap']")
        for s in salary:
            s_avg=s.text
            Salary.append(s_avg)
            S=pd.DataFrame(Salary)
            S=S.rename({0: 'Salary P/M/Y'},axis=1)
        desc=driver.find_elements_by_xpath("//div[@class='summary']/ul/li[1]")
        for d in desc:
            d_name=d.text
            Job_Desc.append(d_name)
            JD=pd.DataFrame(Job_Desc)
            JD=JD.rename({0: 'Job Description'},axis=1)
        postday=driver.find_elements_by_xpath("//span[@class='date ']")
        for p in postday:
            p_day=p.text
            Post_Days.append(p_day)
            PD=pd.DataFrame(Post_Days)
            PD=PD.rename({0: 'Post Days'},axis=1)
       # importing the date and time from the library      
            now=datetime.now()
            crawl_timestamp.append(now.strftime("%H:%M:%S"))
            tim=pd.DataFrame(crawl_timestamp)
            tim=tim.rename({0: 'Time'},axis=1)
            today=date.today()
            crawl_date.append(today.strftime("%d/%m/%Y"))
            dt=pd.DataFrame(crawl_date)
            dt=dt.rename({0: 'Date'},axis=1)
        # now lets scrap the data from the server            
        nxt_button=driver.find_element_by_xpath("//a[@aria-label='Next']")
        nxt_button.click()
        time.sleep(8)

# now lets create the data table and adding the columns into the table
    indeed=pd.concat([JN,LN,CN,S,JD,PD,dt,tim],axis=1)
    indeed.to_csv('Data Analyst indeed', index = True)
    print(indeed)
# Calling Function
data_scientist("https://in.indeed.com/jobs?q=data+scientist&l=Bangalore+City%2C+Karnataka")
```
